```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from Methods.SBM import SBM

models_params = {
    "anvits_model": {
        "n_clusters": 4,
        "scale_embedding": True,
        "n_vectors": -1,
        "dtype": "main",
        "eigen_val_weight": 1/4,
    },
    "spectral_clustering": {
        "n_clusters": 4,
        "scale_embedding": False,
        "n_vectors": 3,
        "dtype": "main",
    },
}

for model_name in models_params:
    model = SBM(**models_params[model_name]).fit()
    models_params[model_name]["labels"] = model.labels
    models_params[model_name]["embeddings"] = model.embedding
```

```python
def label_point(xs, ys, vals, ax):
        for x, y, val in zip(xs, ys, vals):
            ax.text(x+.0002, y, val)

def plot_countries(labels, embeddings):
    xs = embeddings[:, 0]
    ys = embeddings[:, 1]
    label = labels.loc[:, "Label"]

    sns.scatterplot(x = xs, y = ys, c = label)
    # plt.axis('equal')

    label_point(xs, ys, labels.Name, plt.gca())
```

```python
plot_countries(models_params["anvits_model"]["labels"], models_params["anvits_model"]["embeddings"])
```

```python
plot_countries(models_params["spectral_clustering"]["labels"], models_params["spectral_clustering"]["embeddings"])
```
