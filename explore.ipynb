```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from Methods.SBM import SBM
model = SBM(n_clusters=4, scale_embedding=True, n_vectors=2).fit()
sns.heatmap(model.get_country_country(normalize=True), xticklabels=model.country_names, yticklabels=model.country_names)
```

```python
xs = model.embedding[:, 0]
ys = model.embedding[:, 1]
name = model.labels.loc[:, "Name"]
label = model.labels.loc[:, "Label"]

sns.scatterplot(x = xs, y = ys, c = label)
plt.axis('equal')

def label_point(xs, ys, vals, ax):
    for x, y, val in zip(xs, ys, vals):
        ax.text(x+.0002, y, val)

label_point(xs, ys, name, plt.gca())
```
